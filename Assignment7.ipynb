The TIOBE Programming Community index is an indicator of the popularity of programming languages. The index is updated once a month. Use pandas read_html() to fetch HTML tables from [The TIOBE Index Page](https://www.tiobe.com/tiobe-index/). Then write python code with pandas and numpy to answer the subsequent questions. 

```python
url = "http://www.tiobe.com/tiobe-index/"
```

```python
# use pandas read_html to fetch HTML tables
# Your code starts here #
tables = pd.read_html(url, header = 0)
# Your code ends here #
```

Q2. Read the first HTML table into a DataFrame `tiobe`. How many programming languages are there in the DataFrame?

Q3. Delete the `Programming Language` column. Modify the DataFrame in place.

```python
pass
tiobe.drop(columns = ["Programming Language"], inplace = True)
```

```python
tiobe.head()
```

Q4. Change column labels:

- Change `Programming Language.1` to `Programming Language`
- Change the first column label to `Current Ranking`
- Change the second column label to `Last Year`
- Change the last column label to `Rating Change`

Modify the DataFrame in place. Display all column labels.

```python
pass
tiobe = tiobe.rename(columns = {"Programming Language.1":"Programming Language",
                               "Nov 2022" : "Current Ranking", "Nov 2021" : "Last Year",
                                "Change.1" : "Rating Change"})
tiobe
```

Q5. Fill in the missing values in the `Change` column of the `tiobe` DataFrame based on the first two columns: if a language's current ranking is higher than a year ago, then fill in 1; if a language's current ranking is lower than a year ago, then fill in -1; otherwise fill in 0. (Hint: the higher the ranking, the lower the value.) What is the count of 1 on this column?

```python
pass
tiobe["Change"] = np.where(tiobe['Last Year']> tiobe['Current Ranking'],1,
                          np.where(tiobe["Last Year"]< tiobe['Current Ranking'],-1,0))
tiobe.head()
```

```python
tiobe["Change"].value_counts()
```

Q9. Cast the `Ratings` and `Rating Change` columns to the `float32` data type. (Hint: refer to Stackoverflow posts on [converting percent string to float in pandas](https://stackoverflow.com/questions/25669588/convert-percent-string-to-float-in-pandas-read-csv)). Output Java's rating change.

```python
tiobe['Rating Change'] = tiobe['Rating Change'].str.rstrip('%').astype('float') / 100.0
tiobe['Ratings'] = tiobe['Ratings'].str.rstrip('%').astype('float') / 100.0
tiobe
```
